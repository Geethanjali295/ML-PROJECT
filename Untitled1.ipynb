```python
df=df.drop(['Loan_ID'], axis=1)
df
```

```python
df=df.drop(['Gender'], axis=1)
df
```

```python
columns=['Married','Dependents','Self_Employed']
df=df.dropna(subset=columns)
df
```

```python
df.head()
```

```python
df=df.dropna(subset=['Married'])
df
```

```python
# Calculate the mode of the 'Credit_History' column
mode_credit_history = df['Credit_History'].mode()[0]

# Fill missing values in 'Credit_History' with the calculated mode
df['Credit_History'] = df['Credit_History'].fillna(mode_credit_history)

# Verify that there are no more missing values in 'Credit_History'
print(f"Missing values in 'Credit_History' after filling: {df['Credit_History'].isnull().sum()}")

# Display the first few rows to show the updated 'Credit_History' column
display(df.head())
```

```python
df.isnull().sum()
```

```python
# Calculate the median of the 'Loan_Amount_Term' column
median_loan_amount_term = df['Loan_Amount_Term'].median()

# Fill missing values in 'Loan_Amount_Term' with the calculated median
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(median_loan_amount_term)

# Verify that there are no more missing values in 'Loan_Amount_Term'
print(f"Missing values in 'Loan_Amount_Term' after filling: {df['Loan_Amount_Term'].isnull().sum()}")

# Display the first few rows to show the updated 'Loan_Amount_Term' column
display(df.head())
```

```python

median_loan_amount = df['LoanAmount'].median()


df['LoanAmount'] = df['LoanAmount'].fillna(median_loan_amount)


print(f"Missing values in 'LoanAmount' after filling: {df['LoanAmount'].isnull().sum()}")


display(df.head())
```

```python
# Apply log transformation to 'ApplicantIncome' and 'CoapplicantIncome'
# Adding 1 to handle cases where income might be 0, as log(0) is undefined.
df['ApplicantIncome_log'] = np.log(df['ApplicantIncome'] + 1)
df['CoapplicantIncome_log'] = np.log(df['CoapplicantIncome'] + 1)

# Display the head of the DataFrame with the new log-transformed columns
display(df[['ApplicantIncome', 'ApplicantIncome_log', 'CoapplicantIncome', 'CoapplicantIncome_log']].head())

```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
df['Education']=le.fit_transform(df['Education'])
df['Self_Employed']=le.fit_transform(df['Self_Employed'])
df['Married']=le.fit_transform(df['Married'])
```
